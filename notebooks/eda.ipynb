```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,6)
```

```python
df = pd.read_csv('../data/student_lifestyle_dataset.csv')

df.head()
```

```python
df.info()
df.describe()
```

```python
# Check missing values
df.isnull().sum()

# Drop missing (if any)
df = df.dropna()

# Remove duplicates
df = df.drop_duplicates()
```

```python
df.hist(bins=20, figsize=(12,10))
plt.suptitle("Feature Distributions")
plt.show()
```

```python
corr = df.corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.scatterplot(x='sleep_hours', y='gpa', data=df)
plt.title("Sleep vs GPA")
plt.show()
```

```python
sns.scatterplot(x='study_hours', y='gpa', data=df)
plt.title("Study Hours vs GPA")
plt.show()
```

```python
sns.scatterplot(x='screen_time', y='gpa', data=df)
plt.title("Screen Time vs GPA")
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

X = df.drop("gpa", axis=1)
y = df["gpa"]

model = LinearRegression()
model.fit(X, y)

importance = pd.Series(model.coef_, index=X.columns)
importance.sort_values().plot(kind='barh')
plt.title("Feature Importance")
plt.show()
```

```python
fig = px.scatter(df, x="sleep_hours", y="gpa", color="stress_level",
                 title="Sleep vs GPA (Colored by Stress)")
fig.show()
```

```python
df['sleep_category'] = pd.cut(df['sleep_hours'],
                             bins=[0,5,7,10],
                             labels=['Low','Medium','High'])

df.groupby('sleep_category')['gpa'].mean().plot(kind='bar')
plt.title("Average GPA by Sleep Category")
plt.show()
```

```python
sns.boxplot(x='diet_quality', y='gpa', data=df)
plt.title("Diet Quality vs GPA")
plt.show()
```
